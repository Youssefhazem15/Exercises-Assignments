#### #7. Filter the data set. Create three sub DataFrames from the `position` column for `G`, `F`, and `C`.

```python
df1=df[df['pos']=='G']
df2=df[df['pos']=='F']
df3=df[df['pos']=='C']
print(df1.head())
print(df2.head())
print(df3.head())
```

#### #8.  Run `describe()` on these new DataFrames. Compare the mean field goals (the `fg` column) between positions.

```python

describe_G = df1.describe()
describe_F = df2.describe()
describe_C = df3.describe()

mean_fg_G = describe_G['fg']['mean']
mean_fg_F = describe_F['fg']['mean']
mean_fg_C = describe_C['fg']['mean']

# Print the mean field goals for each position
print(f"Mean Field Goals for Guards (G): {mean_fg_G}")
print(f"Mean Field Goals for Forwards (F): {mean_fg_F}")
print(f"Mean Field Goals for Centers (C): {mean_fg_C}")
```
